```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
generador = np.random.default_rng()
```

```python
generador
```

```python
mean = 0.0
sigma = 2.0
length = 1000000
```

```python
sequencex = generador.normal(mean, sigma, length)
```

```python
x = np.arange(length)
```

```python
plt.scatter(x, sequencex)
plt.title('Sequencia aleatoria normal')
plt.show()
```

```python
p, bin_edges = np.histogram(sequencex, 100, density=True)
```

```python
x = (bin_edges[1:]+bin_edges[0:-1])/2.0
```

```python
plt.plot(x, p)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Normal distrib')
plt.show()
```

```python
y = np.arange(length)
```

```python
sequencey = generador.normal(mean, sigma, length)
```

```python
plt.scatter(y, sequencey)
plt.title('Sequencia aleatoria normal')
plt.show()
```

```python
# Dibujar gráficos mixtos
 # Dibujar superficie 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
hist, xedges, yedges = np.histogram2d(sequencex, sequencey, bins=100, range=[[-5, 5], [-5, 5]])

# Construct arrays for the anchor positions of the 16 bars.
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Construct arrays with the dimensions for the 16 bars.
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

plt.show()
```
