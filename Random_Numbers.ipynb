```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
generador = np.random.default_rng()
```

```python
mean = 0.0
sigma = 2.0
length = 1000000
```

```python
sequence = generador.normal(mean, sigma, length)
```

```python
x = np.arange(length)
```

```python
plt.scatter(x, sequence)
plt.title('Sequencia aleatoria normal')
plt.show()
```

```python
p, bin_edges = np.histogram(sequence, 100, density=True)
```

```python
x = (bin_edges[1:]+bin_edges[0:-1])/2.0
```

```python
plt.plot(x, p)
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('Normal distrib')
plt.show()
```
